```python
import numpy as np
import matplotlib.pyplot as plt
import random
from ipywidgets import interact, interactive
from IPython.display import clear_output, display, HTML
```

```python
plt.figure(figsize = (10,10))
theta = np.linspace( 0 , 2 * np.pi , 150 )
plt.grid(True) 
radius = 1
  
a = radius * np.cos( theta )
b = radius * np.sin( theta )
  
plt.plot(a,b)
A = [-1,-1,1,1,-1]
B = [-1,1,1,-1,-1]
plt.plot(A,B)
ax = [0,0]
ay =[-2,2]
plt.plot(ax,ay,color = 'black')
AX =[-2,2]
AY=[0,0]
plt.ylim(-1,1)
plt.xlim(-1,1)
plt.plot(AX,AY,color = 'black')
n=100
inside = 0
i=1
while (i<=n): 
    x = random.uniform(-1,1)
    y = random.uniform(-1,1) 
    if ((x**2)+(y**2))<=1:    
        inside+=1    
        plt.plot(x , y , 'go')  
    else:  
        plt.plot(x , y , 'ro')
    i+=1
pi=(4*inside)/n
print ("The value of pi is:")
print(pi)
plt.title('Monte Carlo simulation for Estimation of Value of PI')
plt.show()
```

```python
def MC(n):
    plt.figure(figsize = (10,10))
    theta = np.linspace( 0 , 2 * np.pi , 150 )
    plt.grid(True) 
    radius = 1
  
    a = radius * np.cos( theta )
    b = radius * np.sin( theta )
  
    plt.plot(a,b)
    A = [-1,-1,1,1,-1]
    B = [-1,1,1,-1,-1]
    plt.plot(A,B)
    ax = [0,0]
    ay =[-2,2]
    plt.plot(ax,ay,color = 'black')
    AX =[-2,2]
    AY=[0,0]
    plt.ylim(-1,1)
    plt.xlim(-1,1)
    plt.plot(AX,AY,color = 'black')
    
    inside = 0
    i=1
    while (i<=n): 
        x = random.uniform(-1,1)
        y = random.uniform(-1,1) 
        if ((x**2)+(y**2))<=1:    
            inside+=1    
            plt.plot(x , y , 'go')  
        else:  
            plt.plot(x , y , 'ro')
        i+=1
    pi=(4*inside)/n
    print ("The value of pi for", n , "points is:")
    print(pi)
    plt.title('Monte Carlo simulation for Estimation of Value of PI')
    plt.show()
```

```python
Plot = interactive(MC, n = [10,100,1000,10000])
display(Plot)
```
