```python
import pandas as pd
import pylab
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("well.xlsx")
df.head()
```

```python
a = df.T
```

```python
a
```

```python
x = a.loc["delta pressure"]
y = a.loc["time"]
z = a.loc["deerivative of pressure"]
```

```python
plt.loglog(y,x, label = "log P")
plt.loglog(y,z , label = "log derivative of P")
# add a label to the x axis
plt.xlabel('Log(time)')

# add a label to the y axis
plt.ylabel('log P and log derivative P')
plt.title('Log Log Plot')

# add a legend 
plt.legend()


```
